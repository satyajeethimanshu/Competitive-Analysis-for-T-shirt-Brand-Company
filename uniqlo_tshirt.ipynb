```python
from selenium import webdriver
import pandas as pd
import requests
driver = webdriver.Chrome()
from bs4 import BeautifulSoup
url="https://www.uniqlo.com/in/en/men/tops/t-shirts"
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
#print(soup.prettify())
```

```python
uniqlo_tshirt=[]
from typing import Text

articals=soup.find('div',class_="fr-product-grid")

for article in articals:
  name=article.find('h2',class_="decscription-text")
  productname=name.get_text()
  name2=article.find('li',class_="fr-status-flag-text")
  productsubname=name2.get_text()
  price=article.find('span',class_="price-original")
  pricelimit=article.find('span',class_="price-limited")
  if price is None:
    productprice=pricelimit.get_text()
  else:
    productprice=price.get_text()
  price2=article.find('div',class_="dual-price-original")
  if price2 is None:
        productactualprice=0
  else:
    productactualprice=price2.get_text()
  rate=article.find('em',class_="fr-implicit")
  if rate is None:
    productrate=0
  else:
    productrate=rate.get_text()
  review=article.find('span',class_="fr-review-count")
  if review is None:
    productreview=0
  else:
    productreview=review.get_text()
  
  uniqlo_tshirt.append([productname,productsubname,productprice,productactualprice,productrate,productreview])
```

```python
df=pd.DataFrame(uniqlo_tshirt,columns=['productname','productsubname','productprice','productactualprice','productrate','productreview'])
df.to_csv('uniqlo_tshirt.csv')
```
