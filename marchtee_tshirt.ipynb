```python
from selenium import webdriver
import pandas as pd
import requests
driver = webdriver.Chrome()
from bs4 import BeautifulSoup
url="https://www.marchtee.com/classics"
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
#print(soup.prettify())
```

```python
marchtee_tshirt=[]
from typing import Text
articals=soup.find_all('div',class_="product-card")
#articals=container.find_all('div',class_="animate-card")

for article in articals:
  name=article.find('p',class_="title footnote")
  productname=name.get_text()
  name2=article.find('p',class_="footnote")
  productsubname=name2.get_text()
  price=article.find('span',class_="price discounted")
  productprice=price.get_text()
  price2=article.find('span',class_="price original strikethrough")
  if price2  is None:
        productactualprice=0
  else:
    productactualprice=price2.get_text()
  
  marchtee_tshirt.append([productname,productsubname,productprice,productactualprice])
```

```python
df=pd.DataFrame(marchtee_tshirt,columns=['productname','productsubname','productprice','productactualprice'])
df.to_csv('marchtee_tshirt.csv')
```
