```python
from selenium import webdriver
import pandas as pd
import requests
driver = webdriver.Chrome()
from bs4 import BeautifulSoup
url="https://www.bewakoof.com/top-wear-for-men"
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
#print(soup.prettify())
```

```python
container=soup.find('div',id="productGrid")
bewakoof_tshirt=[]
from typing import Text
articals=container.find_all('div',class_="plp-product-card")
for article in articals:
  name=article.find('h3')
  productname=name.get_text()
  name2=article.find('h2')
  productsubname=name2.get_text()
  price=article.find('div',class_="discountedPriceText")
  productprice=price.get_text()
  price2=article.find('div',class_="actualPriceText")
  productactualprice=price2.get_text()
  type1=article.find('div',class_="tag_label_plp")
  if type1 is None:
        producttype=0
  else:
        producttype=type1.get_text()
  rate=article.find('span',class_="clr-shade-3")
  if rate is None:
    productrate=0
  else:
    productrate=rate.get_text()
  bewakoof_tshirt.append([productname,productsubname,productprice,productactualprice,producttype,productrate])
```

```python
df=pd.DataFrame(bewakoof_tshirt,columns=['productname','productsubname','productprice','productactualprice','producttype','productrate'])
df.to_csv('bewakoof_tshirt.csv')
```
