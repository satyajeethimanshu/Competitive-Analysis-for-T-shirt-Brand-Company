```python
from selenium import webdriver
import pandas as pd
import requests
driver = webdriver.Chrome()
from bs4 import BeautifulSoup
url="https://www.thesouledstore.com/men/t-shirts"
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
#print(soup.prettify())
```

```python
thesouledstore_tshirt=[]
from typing import Text
container=soup.find('div',class_="container plisting fullWidth")
articals=container.find_all('div',class_="animate-card")

for article in articals:
  name=article.find('h5',class_="text-left")
  productname=name.get_text()
  name2=article.find('div',class_="listprice ecltext")
  productsubname=name2.get_text()
  price=article.find('span',class_="offer fsemibold")
  productprice=price.get_text()
  price2=article.find('span',class_="product-price")
  if price2  is None:
        productactualprice=0
  else:
    productactualprice=price2.get_text()
  soldout=article.find('div',class_="souled-out-image-blur")
  if soldout is None:
    productsoldout=0
  else:
    productsoldout=soldout.get_text()
  
  thesouledstore_tshirt.append([productname,productsubname,productprice,productactualprice,productsoldout])
```

```python
df=pd.DataFrame(thesouledstore_tshirt,columns=['productname','productsubname','productprice','productactualprice','productsoldout'])
df.to_csv('thesouledstore_tshirt.csv')
```
